```python
class Car:
  carCount = 0
  def __init__(self, name, color):
    self.name = name
    self.color = color
    Car.carCount += 1
  def getCar(self):
    return "Name: " + self.name + ", Color: " + self.color

ford = Car("Ford", "Merah")
toyota = Car("Toyota", "Hijau")
honda = Car("Honda", "Hitam")
bmw = Car("BMW", "Biru")
print("Total mobil: %d" % Car.carCount)
print("Mobil 1 \n" + ford.getCar())
print("Mobil 2 \n" + toyota.getCar())
print("Mobil 3 \n" + honda.getCar())
print("Mobil 4 \n" + bmw.getCar())
```

```python
# Daftar angka
angka = [1, 2, 3, 4, 5]

# Menggunakan map untuk menerapkan fungsi kuadrat ke setiap elemen dalam daftar angka
hasil = map(lambda x:x**2, angka)

# Mengonversi hasil map menjadi list dan mencetaknya
print(list(hasil))  # Output: [1, 4, 9, 16, 25]
```

```python
hewan = ['kucing', 'burung', 'anjing', 'kelinci']
angka = [4, 2, 8, 2]

# Menghitung banyaknya elemen pada list hewan menggunakan fungsi len
print(f"len hewan = {len(hewan)}")

# Mengetahui index pada elemen tertentu menggunakan method index
print(f"index hewan = {hewan.index('burung')}")

# Menghitung banyaknya elemen tertentu menggunakan method count
print(f"count angka 2 = {angka.count(2)}")

# Mengurutkan elemen pada list dengan fungsi sort
hewan.sort()
print(f"hewan.sort = {hewan}")

# Mengurutkan elemen secara descending
hewan.sort(reverse = True)
print(f"hewan reverse sort = {hewan}")

# Menggabungkan menggunakan operator +
print(f"hewan + angka = {hewan + angka}")

# Mereplikasi menggunakan operator *
print(f"angka x 2 = {angka * 2}")
```

```python
# In trigonometry, angles can be measured in radians. One complete revolution around a circle is 2 radians, equivalent to 360 degrees.
# The sine function (sin) is a trigonometric function that relates an angle of a right triangle to the ratios of two of its sides.

# Consider the variable x as being in the interval [0, 1]. To convert this interval into one full cycle of a sine wave, scale x by 2:
# - When x = 0, 2 * 0 = 0 radians (0 degrees)
# - When x = 1, 2 * 1 = 2 radians (360 degrees)

# This scaling is used to generate a sine wave over one complete cycle.
# Let's break down the given Python code:

import numpy as np  # Import NumPy, a powerful numerical computing library

x = np.linspace(0, 1, 100)  # Generate 100 linearly spaced values from 0 to 1
s = np.sin(2 * np.pi * x)   # Compute the sine of 2 times each value in x

# 's' now contains the sine of each corresponding value in 'x', producing an array of sine values that complete one full oscillation.
```

```python
from sklearn.preprocessing import LabelBinarizer

# Sample data
y = [1, 2, 3, 4]

# Initialize and fit LabelBinarizer
lb = LabelBinarizer()
lb.fit(y)

# Transform the data
y_new = lb.transform([1])
print(y_new)  # Transformed array

# Inverse transform
original_class = lb.inverse_transform(y_new)
print(original_class)  # Should output [1]
```
